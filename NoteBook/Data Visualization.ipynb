```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the DataFrame with predictions
result_df = pd.read_csv(r"C:\Users\olive\Downloads\0_DONE_FINAL_1.csv")

# Plot the distribution of predicted barangays
plt.figure(figsize=(12, 8))
sns.countplot(data=result_df, x='barangay', order=result_df['barangay'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Distribution of Barangays')
plt.xlabel('barangay')
plt.ylabel('Count')
plt.show()
```

```python

# Create a scatter plot without the legend
plt.figure(figsize=(10, 8))
sns.scatterplot(x='cus_longitude', y='cus_latitude', hue='barangay', data=result_df, palette='viridis', alpha=0.7, legend=False)

# Plot labels and title
plt.xlabel('Customer Longitude')
plt.ylabel('Customer Latitude')
plt.title('Geographical Distribution of Customers by Barangay')
plt.grid(True)

# Show plot
plt.show()
```
